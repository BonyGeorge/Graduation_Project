```python
import cv2
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
vidcap = cv2.VideoCapture('/content/gdrive/MyDrive/bunny.mp4')
success, image = vidcap.read()
count = 0
while success:
  cv2.imwrite("/content/gdrive/MyDrive/bunny_images/frame%d.jpg" % count, image)     # save frame as JPEG file      
  success,image = vidcap.read()
  print('Read a new frame: ', success)
  count += 1
```
